# RFM Analysis

Recency, frequency, monetary value (RFM) is a marketing analysis tool used to identify a firm's best clients based on the nature of their spending habits.    

Recency (R):   How long ago they made a purchase       
Frequency (F): How often they make purchases       
Monetary (M): How much money they spend       

```python
# Extracting date from InvoiceDate column

data['Date'] = data.InvoiceDate.dt.date
now = dt.date(2022, 10, 18)
```

```python
data.head()
```

```python
# Revenue per month for year 2010 and 2011

date_Revenue = pd.DataFrame(data.Revenue.groupby(data.Date).sum()).reset_index()
date_Revenue['year_month'] = pd.to_datetime(date_Revenue['Date']).dt.strftime('%Y-%m')

date_Revenue = pd.DataFrame(date_Revenue.Revenue.groupby(date_Revenue.year_month).sum()).reset_index()

plt.figure(figsize = (12,6))
lineplot = sns.lineplot(x = date_Revenue.year_month, y = date_Revenue.Revenue)
lineplot.set(xticks = date_Revenue.year_month.values)
plt.title('Revenue per month for all countries')
plt.show()
```

Revenue increased drastically from August 2011 to November 2011 and suddenly dropped in the next month

## Recency

To get Recency we need to calculate total number of days passed between today and the date of last transaction each customer 

```python
# Recency feature

rfm = pd.DataFrame()
data1 = ((now - data.Date).dt.days).groupby(data.CustomerID).max()
rfm['Recency'] = data1
```

```python
rfm.head()
```

## Frequency

To get frequency we need to find the total number of transactions for each customer

```python
# Frequency feature

data2 = data.drop_duplicates(['InvoiceNo','CustomerID'])
rfm['Frequency'] = (data2.InvoiceNo.groupby(data2.CustomerID).count())
rfm.head()
```

## Monetary

Total revenue generated from each customer

```python
# Monetary feature

data3 = (data.Quantity * data.UnitPrice).groupby(data.CustomerID).sum()
rfm['Monetary'] = data3
rfm.head()
```

We will compare Monetary Score and Recency Score to get the RFM score

```python
rfm.head()
```

```python
# Segmenting customers based on RFM score

rfm.Customer_behaviour.loc[rfm.Customer_behaviour == '51'] = 'Price_Sensitive'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '15') | (rfm.Customer_behaviour == '25')] = 'Cant_Lose'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '11') |  (rfm.Customer_behaviour == '21') |
                          (rfm.Customer_behaviour == '12') |  (rfm.Customer_behaviour == '22')] = 'Lost'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '13') |  (rfm.Customer_behaviour == '14')
                          | (rfm.Customer_behaviour == '23') |  (rfm.Customer_behaviour == '24')] = 'Hibernating'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '31') |  (rfm.Customer_behaviour == '32')] = 'About_to_sleep'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '33')] = 'Needs_Attention'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '34') |  (rfm.Customer_behaviour == '35') |
                           (rfm.Customer_behaviour == '44') |  (rfm.Customer_behaviour == '45')] = 'Loyal Customers'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '41')] = 'Promising'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '42') |  (rfm.Customer_behaviour == '43')
                          | (rfm.Customer_behaviour == '53')] = 'Potential_Loyalist'

rfm.Customer_behaviour.loc[(rfm.Customer_behaviour == '54') |  (rfm.Customer_behaviour == '55')] = 'Champions'

rfm.Customer_behaviour.loc[rfm.Customer_behaviour == '52'] = 'Recent_Users'
```

# Clustering

Before clustering lets check for distribution of the features

```python
from scipy.stats import skew, kurtosis
figure, ax = plt.subplots(3, 1, figsize = (10,10))
ax = ax.flatten()

x = rfm[['Recency', 'Frequency', 'Monetary']]

index = 0
for i in x.columns:
    plot = sns.histplot(data = x[i], kde = 1, ax = ax[index])
    print(f'Kurtosis {i}: ',kurtosis(x[i]))
    print(f'Skewness {i}: ',skew(x[i]))
    index += 1
```

Kurtosis value for normal distribution ranges between -2 and 2.          
Skewness value for normal distribution is 0.

### Log Transformation

Recency feature seems to be close to normal distribution.     
However, Frequency and Monetary are highly right skewed with high kurtosis values.     
Applying log transformation to Frequency and Monetary to make them normal

```python
x['Recency_log'] = np.log(x.Recency)
x['Frequency_log'] = np.log(x.Frequency)
x['Monetary_log'] = np.log(x.Monetary)
x.head()
```

# Elbow Curve

```python
# Plotting Elbow Curve to obtain n_clusters

x_subset = x[['Recency_log', 'Frequency_log', 'Monetary_log']] # Creating a subset

inertia_value = []

for i in range(1,9):
    kmean = KMeans(n_clusters = i, random_state = 0)
    kmean.fit(x_subset)
    inertia_value.append(kmean.inertia_)

plt.figure(figsize = (12,6))
elbow_curve = sns.lineplot( x = range(1,9), y = inertia_value, markers = 'o', size = 20)

plt.xlabel('No. of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Curve')
plt.grid()
plt.show()
```

# Checking for silhouette score to check the cluster formation 

```python
for i in range(2,9):
    kmean1 = KMeans(n_clusters = i, random_state = 0)
    cluster_labels = kmean1.fit_predict(x_subset)

# The silhouette_score gives the average value for all the samples.
# This gives a perspective into the density and separation of the formed clusters

    silhouette_avg = silhouette_score(x_subset, cluster_labels)
    print(f'For n_clusters = {i} the average silhouette_score is : {np.round(silhouette_avg * 100, 2)} %')
```

Based on silhouette score and elbow curve we will go with 4 clusters having low inertia

```python
# Kmeans Clustering for 4 clusters
# Random seed = 0

model = KMeans(n_clusters = 4, random_state = 0)
cluster_labels = model.fit_predict(x_subset)
x['clusters'] = cluster_labels
x_subset['clusters'] = cluster_labels
x.head()
```

```python
x.groupby(x.clusters).mean()
```

```python
# ScatterPlot 

sns.scatterplot( x = x.Monetary_log, y = x.Frequency_log, hue = x.clusters, palette = 'tab10')
```

## Cross Verifying the formation of clusters using different Classification Models

```python
from sklearn.model_selection import train_test_split

# Splitting into X and Y 
X = x_subset.loc[:, x_subset.columns != 'clusters']
Y = x_subset.clusters

# Splitting into train test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 0, stratify = Y) 
```

```python
# KNN Classifier

knn = KNeighborsClassifier(n_neighbors = 4)

knn.fit(X_train, Y_train)
knn_predict = knn.predict(X_test)

print('Accuracy Score using KNN: ',np.round(accuracy_score(knn_predict, Y_test)*100, 2), '%')
print('\n')
print(confusion_matrix(Y_test, knn_predict))
print('\n')
print(classification_report(Y_test, knn_predict)) 
```

```python
# Logistic Regression

lr = LogisticRegression(max_iter = 1000, random_state = 0)

lr.fit(X_train, Y_train)
lr_predict = lr.predict(X_test)

print('Accuracy Score using Logistic Regression: ', np.round(accuracy_score(lr_predict, Y_test)*100, 2), '%')
print('\n')
print(confusion_matrix(Y_test, lr_predict))
print('\n')
print(classification_report(Y_test, lr_predict))
```

```python
# NaiveBayes Classifier (Gaussian Classifier)

gnb = GaussianNB()

gnb.fit(X_train, Y_train)
gnb_predict = gnb.predict(X_test)

print('Accuracy Score using Naive Bayes: ', np.round(accuracy_score(gnb_predict, Y_test)*100, 2), '%')
print('\n')
print(confusion_matrix(Y_test, gnb_predict))
print('\n')
print(classification_report(Y_test, lr_predict))
```

All the 3 classifiers KNN Classifier, Logistic Regresion and GaussianNB were used where target variables were considered as clusters we get using KMeans algorithm. We get very good test accuracy which proves that the clusters predicted by KMeans are correct.

# Association Rules

Association rule learning is a rule-based machine learning method for discovering interesting relations between variables in large databases.      
We can use association rules to compare different buying patterns of customers which will help us recommend products to customers which will help increase customer satification and involvement and also increase Revenue

```python
# Extracting required features 

dataset = data[['Description','CustomerID']]
dataset.head()
```

```python
top_items = dataset.Description.value_counts().head(400).index
percent_of_total = np.round((dataset.Description.value_counts().head(400).values.sum() / dataset.Description.shape[0]), 2)
percent_of_total
```

```python
# Combining the products for each Customers 

dataset = (dataset.Description.groupby(data.CustomerID).agg([('Description', ', '.join)])).reset_index()
dataset.head()
```

```python
for i in dataset.Description:
    for j in top_items:
        
        if j in i:
            dataset[j] = 1
        else:
            dataset[j] = 0
```

```python
dataset.head()
```

```python
# Dropping columns

dataset.drop(columns = ['CustomerID', 'Description'], axis = 1, inplace = True)
```

```python
from mlxtend.frequent_patterns import apriori, association_rules

frequent_products = apriori(dataset, min_support = 0.01, use_colnames = True, max_len = 3)
frequent_products = frequent_products.sort_values('support', ascending = False)
frequent_products
```

```python
# Association rules of the dataset

rules = association_rules(frequent_products, metric = 'lift', min_threshold = 1)
rules = rules.sort_values('lift', ascending=False)
rules.head()
```

```python
# Removing redundant itemsets

def to_list(i):
    return sorted(list(i))

red_data = rules.antecedents.apply(to_list) + rules.consequents.apply(to_list)
red_data = red_data.apply(sorted)
rules_sets = list(red_data)
rules_sets
```

```python
# Getting unique rules 

unique_rules_set = [list(m) for m in set(tuple(i) for i in rules_sets)]
unique_rules_set
```

```python
# Retrieving index of unique itemsets

index_rules = []

for i in unique_rules_set:
    index_rules.append(rules_sets.index(i))
    
rules_no_red = rules.iloc[index_rules,:]
rules_no_red = rules_no_red.sort_values('lift', ascending= False)
rules_no_red
```

```python
# Defining a recommendation system for the customer

def recommend(item):
    to_recommend = []
    for i in range(len(rules_no_red.antecedents)):
        if item in rules_no_red.loc[rules_no_red.index[i],'antecedents']:
            to_recommend.extend(list(rules_no_red.loc[rules_no_red.index[i],'consequents']))
            
    if to_recommend == []:
        print('There is no product to recommend!')
    else:
        print('We recommend you to check out these products: ')
        return set(to_recommend)
```

```python
# Testing the recomend function

recommend('DOUGHNUT LIP GLOSS')
```
